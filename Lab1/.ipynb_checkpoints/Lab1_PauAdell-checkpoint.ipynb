Consider the cube centered at the origin and whose edges have length 2 and are axis parallel. 

```python
# Origin
origin = vector((0,0,0))
# Canonical axes
axis1=vector((1,0,0))
axis2=vector((0,1,0))
axis3=vector((0,0,1))

# Graphic representation, in order to plot them
canonical_axes = arrow(origin, origin + axis1, color='green') + \
    arrow(origin, origin + axis2, color='red') + \
    arrow(origin, origin + axis3, color='blue')
```

```python
# Basic cube definition with lenght 2 edges
#              0                          1                 2                  3                4                  5                 6                7
vertices = [vector((-1,-1,-1)),vector((1,-1,-1)),vector((-1,1,-1)),vector((-1,-1,1)), vector((1,1,-1)), vector((1,-1,1)), vector((-1,1,1)) ,vector((1,1,1))]
faces = [[0,2,4,1], [0, 2, 6,3], [0, 3, 5, 1], [5, 1, 4, 7], [4,2,6,7], [6,7,5,3]]
polygons3d(faces, vertices, alpha=0.3)+canonical_axes
```

Let
us truncate the cube in the following way: consider its vertex v = (1, 1, 1), compute the midpoints
of all edges incident to v, and replace v by the new triangular face resulting from connecting the
midpoints.

```python
v = vertices[7]
print(v)
incidentVectors = [vertices[4], vertices[6], vertices[5]]
incidentVectors
```

```python
# Compute midpoints of all edges
midPoints = [(v + aux)/ 2 for aux in incidentVectors]
midPoints
```

We will call the resulting heptahedron a cropped cube. It is illustrated in the figure
below.

```python
# Cropped cube with new vertices
croppedV = [vertices[0], vertices[1], vertices[2], vertices[3], vertices[4], vertices[5], vertices[6], midPoints[0], midPoints[1], midPoints[2]]
croppedF = [[0,2,4,1], [0,2,6,3], [0,3,5,1], [5,1,4,7,9], [4,2,6,8,7], [6,8,9,5,3], [7,8,9]]
p = Graphics()
p += polygons3d(croppedF, croppedV, alpha=0.3)+canonical_axes
for m in midPoints:
    p += point3d(m, color='yellow', size=10)
p.show()
```

Consider a vertical unit length segment, s1, starting at the origin. Place and visualize a unit
length segment s2 on top of the center of the triangular face of C1, and perpendicular to it.
Compute s2 from s1 by changing coordinate systems.

With the 3 points we can get the perpedicular (the normal of the plane) to the plane as we can create 2 vectors and in order to get to the center we can do the midpoint between the 3 points as following:

```python
# First visualize only the face with the points
v = [midPoints[0], midPoints[1], midPoints[2]]
f = [[0,1,2]]
p = Graphics()
p += polygons3d(f, v, alpha=0.3)+canonical_axes
for m in midPoints:
    p += point3d(m, color='green', size=5)
p.show()
```

```python
# Lets get the middle points of the edges
triangleMidPoints = [(midPoints[i-1] + aux)/ 2 for i,aux in enumerate(midPoints)]
triangleMidPoints

p = Graphics()
p += polygons3d(f, v, alpha=0.3)+canonical_axes
for m in triangleMidPoints:
    p += point3d(m, color='red', size=5)
p.show()
```

```python
# Lets get MA, MB
MA = (midPoints[0]-triangleMidPoints[2])
MB = (midPoints[1]-triangleMidPoints[0])

MA_MB_axes = arrow(triangleMidPoints[2], triangleMidPoints[2] + MA, color='yellow') + \
    arrow(triangleMidPoints[0], triangleMidPoints[0] + MB, color='gray')

polygons3d(f, v, alpha=0.3)+MA_MB_axes
```

```python
# Intersection between MA, MB is the centroid (center of the triangle)
# Parametric equations of the lines
L1 = midPoints[0] + var('t') * MA  # Parametric line 1
L2 = midPoints[1] + var('s') * MB  # Parametric line 2

# Solve the system of equations to find t and s
eqs = [L1[i] == L2[i] for i in range(3)]
solution = solve(eqs, var('t, s'))

# Substitute the parameter t or s into one of the parametric lines to get the intersection point
intersection = L1.subs(solution[0])
intersection
```

```python
p = Graphics()
p += polygons3d(f, v, alpha=0.3)+canonical_axes
p += point3d(intersection, color='yellow', size=10)
p.show()
```

Now that we have the intersection or midle of the triangle lets get the normal of the plane:

```python
# We just need to calculate the cross product of two vectors of the plane which we should already have.
v = vector(midPoints[0] - midPoints[1])
w = vector(midPoints[0] - midPoints[2])

normal_triangle_plane = v.cross_product(w);

new_axes = arrow(vector((0,0,0)), normal_triangle_plane, color='yellow')
normal_triangle_plane
```

```python
p = Graphics()
p += canonical_axes
p += new_axes
p.show()
```

Lets recap, we have the vertical unit segment S1, the origin position of where it will be placed in
the new coordinate system W, and we need the find the changing basis matrix:

```python
S1 = vector((0,0,1))
W = intersection

# Changing basis matrix:
# e1' = vector that goes through intersection (and normalize)
# e2' = e1'x e2'
# e3' = normal_triangle_plane (normalized)
```

Lets get the value of the chaning basis matrix:

```python
e1 = MA/MA.norm()
e3 = normal_triangle_plane/normal_triangle_plane.norm()
e2 = e1.cross_product(e3)

# Origin
NOrigin = intersection

# Graphic representation, in order to plot them
new_axes = arrow(NOrigin, NOrigin + e1, color='green') + \
    arrow(NOrigin, NOrigin + e2, color='red') + \
    arrow(NOrigin, NOrigin + e3, color='blue')

p = Graphics()
p += new_axes
p += polygons3d(croppedF, croppedV, alpha=0.3)#+canonical_axes
p.show()
```

```python
T = Matrix([
    [e1[0], e2[0], e3[0], W[0]],
    [e1[1], e2[1], e3[1], W[1]],
    [e1[2], e2[2], e3[2], W[2]],
    [    0,     0,     0,    1]
])
S2 = T * vector((S1[0], S1[1], S1[2], 1))
S2 = vector((S2[0], S2[1], S2[2]))
S2 = S2/S2.norm()
S2
```

By changing coordinate systems to C1 , place and visualize a copy C2 of C1 with its
triangular face facing the floor (i.e., the plane z = 0 in SageMath), and above the floor (i.e.,
all z-coordinates of C2 need to be non-negative).

Lets get the rotation of each of the axis from one coordinate system from the other one, lets remember which
coordinates systems we have:

```python
C2x = MA/MA.norm()
C2z = normal_triangle_plane/normal_triangle_plane.norm()
C2y = C2x.cross_product(C2z)
```

```python
A = Matrix([[C2x.dot_product(axis1), C2x.dot_product(axis2), C2x.dot_product(axis3)],
            [C2y.dot_product(axis1), C2y.dot_product(axis2), C2y.dot_product(axis3)],
            [C2z.dot_product(axis1), C2z.dot_product(axis2), C2z.dot_product(axis3)]])
A
```

```python
center = vector((0,0,0))
NW = vector((0, 0, -(center-intersection).norm()))
```

```python
T1 = Matrix([
    [A[0][0], A[0][1], A[0][2], NW[0]],
    [A[1][0], A[1][1], A[1][2], NW[1]],
    [A[2][0], A[2][1], A[2][2], NW[2]],
    [0, 0, 0, 1]
])
T1
```

```python
C2_inverted = []

for v in croppedV:
    oldP = vector((v[0], v[1], v[2], 1))
    newP = T1 * oldP
    C2_inverted.append(vector((newP[0],newP[1],newP[2])))

p = Graphics()
p += polygons3d(croppedF, C2_inverted, alpha=0.3)+canonical_axes
p.show()
```

```python
C2z = -normal_triangle_plane/normal_triangle_plane.norm() # change z direction

A2 = Matrix([[C2x.dot_product(axis1), C2x.dot_product(axis2), C2x.dot_product(axis3)],
            [C2y.dot_product(axis1), C2y.dot_product(axis2), C2y.dot_product(axis3)],
            [C2z.dot_product(axis1), C2z.dot_product(axis2), C2z.dot_product(axis3)]])

NW = vector((0, 0, (center-intersection).norm()))

T2 = Matrix([
    [A2[0][0], A2[0][1], A2[0][2], NW[0]],
    [A2[1][0], A2[1][1], A2[1][2], NW[1]],
    [A2[2][0], A2[2][1], A2[2][2], NW[2]],
    [0, 0, 0, 1]
])

C2 = []

for v in croppedV:
    oldP = vector((v[0], v[1], v[2], 1))
    newP = T2 * oldP
    C2.append(vector((newP[0],newP[1],newP[2])))

p = Graphics()
p += polygons3d(croppedF, C2, alpha=0.3)+canonical_axes
p.show()
```

Without moving C1 from its position, attach to C1 a third copy C3 of the cropped cube,
so that their triangular faces match. Visualize the result. As before, solve this problem by
changing coordinates systems.

```python
C2z = normal_triangle_plane/normal_triangle_plane.norm() # change z direction

A3 = Matrix([[axis1.dot_product(C2x), axis1.dot_product(C2y), axis1.dot_product(C2z)],
            [axis2.dot_product(C2x), axis2.dot_product(C2y), axis2.dot_product(C2z)],
            [axis3.dot_product(C2x), axis3.dot_product(C2y), axis3.dot_product(C2z)]])

T2 = Matrix([
    [A3[0][0], A3[0][1], A3[0][2], intersection[0]],
    [A3[1][0], A3[1][1], A3[1][2], intersection[1]],
    [A3[2][0], A3[2][1], A3[2][2], intersection[2]],
    [0, 0, 0, 1]
])

C3 = []

for v in C2:
    oldP = vector((v[0], v[1], v[2], 1))
    newP = T2 * oldP
    C3.append(vector((newP[0],newP[1],newP[2])))

p = Graphics()
p += polygons3d(croppedF, croppedV, alpha=0.3)#+canonical_axes
p += polygons3d(croppedF, C3, alpha=0.3)+canonical_axes
p.show()
```
